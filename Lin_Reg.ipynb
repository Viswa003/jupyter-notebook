```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
data=pd.read_csv("iris.csv")
data.plot(kind='scatter',x='Age',y='BP')
reg1='Age~BP'
reg1output=smf.ols(reg1,data).fit()
print(reg1output.summary())
```
